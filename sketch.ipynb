```python
import os
from pathlib import Path
```

```python
def calculate_average_image_size(folder_path):
    # Supported image file extensions
    image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp'}
    
    total_size_bits = 0
    image_count = 0
    
    # Iterate through all files in the folder
    for file_path in Path(folder_path).iterdir():
        if file_path.is_file() and file_path.suffix.lower() in image_extensions:
            try:
                # Get file size in bytes and convert to bits
                file_size_bytes = os.path.getsize(file_path)
                file_size_bits = file_size_bytes * 8
                
                total_size_bits += file_size_bits
                image_count += 1
                
                print(f"Processed: {file_path.name} - Size: {file_size_bits:,} bits")
            except Exception as e:
                print(f"Error processing {file_path.name}: {e}")
    
    if image_count == 0:
        print("No image files found in the specified folder.")
        return None
    
    average_size_bits = total_size_bits / image_count
    return average_size_bits

# Example usage
# folder_path = "/home/fe/xie/datasets/dummy/valid"  # Replace with your folder path
folder_path = "/home/fe/xie/AutoencoderCompression_pytorch/outputs/reconstruction"  # Replace with your folder path
average_size = calculate_average_image_size(folder_path)

if average_size is not None:
    print(f"\nAverage image size: {average_size:,.2f} bits")
    print(f"({average_size/8:,.2f} bytes)")
    print(f"({average_size/(8 * 1024):,.2f} kilobytes)")
    print(f"({average_size/(8 * 1024 * 1024):,.2f} megabytes)")
```

```python
import os
from pathlib import Path

def calculate_average_pth_size(folder_path):
    # Initialize counters
    total_size_bits = 0
    file_count = 0
    
    # Iterate through all files in the folder
    for file_path in Path(folder_path).iterdir():
        if file_path.is_file() and file_path.suffix.lower() == '.pth':
            try:
                # Get file size in bytes and convert to bits
                file_size_bytes = os.path.getsize(file_path)
                file_size_bits = file_size_bytes * 8
                
                total_size_bits += file_size_bits
                file_count += 1
                
                print(f"Processed: {file_path.name} - Size: {file_size_bits:,} bits")
            except Exception as e:
                print(f"Error processing {file_path.name}: {e}")
    
    if file_count == 0:
        print("No .pth files found in the specified folder.")
        return None
    
    average_size_bits = total_size_bits / file_count
    return average_size_bits

# Example usage
folder_path = "/home/fe/xie/AutoencoderCompression_pytorch/outputs/binary"  # Replace with your folder path
average_size = calculate_average_pth_size(folder_path)

if average_size is not None:
    print(f"\nAverage .pth file size: {average_size:,.2f} bits")
    print(f"({average_size/8:,.2f} bytes)")
    print(f"({average_size/(8 * 1024):,.2f} kilobytes)")
    print(f"({average_size/(8 * 1024 * 1024):,.2f} megabytes)")
```
